```python
import fiona
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection
from descartes import PolygonPatch
```

```python
from shapely.geometry import MultiPolygon, shape
```

```python
%matplotlib nbagg
```

```python
fig, ax = plt.subplots()
with fiona.open('R02_11_WGS84/R02_WGS84.shp', 'r') as features:
    mp = MultiPolygon([shape(pol['geometry']) for pol in fiona.open('R02_11_WGS84/R02_WGS84.shp')])
    minx, miny, maxx, maxy = mp.bounds
    w, h = maxx - minx, maxy - miny
    ax.set_xlim(minx - 0.2 * w, maxx + 0.2 * w)
    ax.set_ylim(miny - 0.2 * h, maxy + 0.2 * h)
    ax.set_aspect(1)
    patches = []
    for idx, p in enumerate(mp):
        patches.append(PolygonPatch(p, fc="gray", ec='black', alpha=1, zorder=1))
    ax.add_collection(PatchCollection(patches, match_original=True))
    plt.show()
```
